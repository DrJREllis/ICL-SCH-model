```python
import pandas as pd
import numpy as np

filename = "Test"

prevalence = pd.read_json('outputs/' + filename + '_prevalences.json')
targets = pd.read_json('outputs/' + filename + '_targetProbabilities.json')
params = np.load('outputs/' + filename + '_params.npy',allow_pickle=True).item()

# eggCounts = np.load('outputs/' + filename + '_eggCounts.npy')
# wormCounts = np.load('outputs/' + filename + '_wormCounts.npy')
# ages = np.load('outputs/' + filename + '_ages.npy')
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

plt.style.use('ggplot')

tArray = prevalence.Time - np.min(prevalence.Time) 

tStep=2

plt.plot(tArray[::tStep] ,prevalence.iloc[::tStep,1],label='PSAC', color='blue')
plt.plot(tArray[::tStep] ,prevalence.iloc[::tStep,4],label='SAC', color='r')
plt.plot(tArray[::tStep] ,prevalence.iloc[::tStep,7],label='Adults', color='g')
plt.plot(tArray[::tStep] ,prevalence.iloc[::tStep,10],label='PSAC high intensity',linestyle='dotted', color='blue')
plt.plot(tArray[::tStep] ,prevalence.iloc[::tStep,13],label='SAC high intensity',linestyle='dotted', color='r')
plt.plot(tArray[::tStep] ,prevalence.iloc[::tStep,16],label='Adults high intensity',linestyle='dotted', color='g')


plt.plot(tArray[::tStep] ,np.full(len(tArray[::tStep]),0.02),'--',color=[0.5,0.5,0.5])

plt.text(15.2,0.025,'2% threshold')


plt.ylabel('Prevalence')
plt.xlabel('Year')
plt.legend(ncol=2)
# plt.xlim(0,20)

plt.show()     
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

plt.style.use('ggplot')


tArray = targets.Time - np.min(targets.Time) + 1


tStep = 10

for i in range(1,np.size(targets,axis=1)):
    plt.plot(tArray[::tStep] ,targets.iloc[::tStep,i]*100,label=targets.columns[i])


plt.ylabel('Probability of target')
plt.xlabel('Year')
plt.legend(ncol=1)

# plt.title('Probability of reaching target')

plt.show()     
```
